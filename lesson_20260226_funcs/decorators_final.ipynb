What Are Decorators in Python?
A decorator in Python is a design pattern used to modify or extend the behavior of functions or methods without altering their actual code. They allow for reusable, clean, and readable enhancements to functions.

Core Concept
A decorator takes a function as input and returns a new function with added behavior.
Decorators use Python’s ability to treat functions as first-class objects (This means you can:

Assign them to variables.
Pass them as arguments to other functions.
Return them from functions.
Store them in data structures.)

```python
def greet():
    print("Hello, World!")
```

```python
greet()
```

```python
def my_decorator(func):
    def wrap():
        print("Before the function call")
        func()
        print("After the function call")
    return wrap
```

```python
greet_decorated = my_decorator(greet)
greet_decorated()
```

```python
@my_decorator
def greet_decorated_v2():
    print("Hello, World!")
```

```python
greet_decorated_v2()
```

Why is this useful?

Example: build a logging decorator to output informations about a given function. Use a sum function a simple sum function

```python
def add(a,b):
    return a + b
```

```python
def log_info(func):
    def wrapper(*args, **kwargs):
        print(f"Function {func.__name__} called with {args} and {kwargs}")
        result = func(*args, **kwargs)
        print(f"Function {func.__name__} returned {result}")
        return result
    return wrapper
```

```python
@log_info
def decorated_add(a,b):
    return a + b
```

```python
add(3,2)
```

```python
decorated_add(3,2)
```

Exercise: enhance the decorator logging also the time of execution of the function

```python
def log_time(func):
    import time
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Function {func.__name__} took {end-start} seconds")
        return result
    return wrapper
```

```python
@log_time
@log_info
def decorated_add_v2(a,b):
    return a + b
```

```python
decorated_add_v2(3,2)
```
