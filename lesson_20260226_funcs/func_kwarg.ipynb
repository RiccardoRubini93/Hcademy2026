```python
list1 = [1, 2]
list2 = [3, 4]
combined = [*list1, *list2]

print(combined)  # Output: [1, 2, 3, 4]
```

### Unione di due liste con l'operatore *
Questa cella mostra come unire due liste (`list1` e `list2`) utilizzando l'operatore di unpacking `*`. Il risultato è una nuova lista che contiene tutti gli elementi delle due liste originali.

```python
numbers = [1, 2, 3, 4, 5]

a, b, *rest = numbers
print(a)     # Output: 1
print(b)     # Output: 2
print(rest)  # Output: [3, 4, 5]

# Collecting elements at the end
*start, x, y = numbers
print(start)  # Output: [1, 2, 3]
print(x)      # Output: 4
print(y)      # Output: 5
```

### Unpacking di liste in variabili e raccolta di elementi restanti
Questa cella mostra come estrarre i primi due elementi di una lista in variabili (`a`, `b`) e raccogliere i restanti in una nuova lista (`rest`) usando l'operatore `*`. Mostra anche come raccogliere tutti gli elementi tranne gli ultimi due in una lista (`start`) e assegnare gli ultimi due a variabili separate (`x`, `y`).

```python
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 4,"c": 3}
merged_dict = {**dict1, **dict2}  # Merge dictionaries
print(merged_dict)  # Output: {'a': 1, 'b': 2, 'c': 3}
```

### Merge di dizionari con l'operatore **
Questa cella mostra come unire due dizionari (`dict1` e `dict2`) usando l'operatore di unpacking `**`. In caso di chiavi duplicate, il valore del secondo dizionario sovrascrive quello del primo.

```python
def sum_numbers(*args):
    print(f"args is of type: {type(args)} e conents: {args}")
    total = sum(args)  # Sum all the positional arguments
    print(f"The sum of the numbers is: {total}")

sum_numbers(1, 2, 3, 4, 5)
# Output:
# The sum of the numbers is: 15
```

### Funzione con *args per sommare numeri
Questa cella definisce una funzione (`sum_numbers`) che accetta un numero variabile di argomenti posizionali usando `*args`. La funzione stampa il tipo di `args` (una tupla) e la somma di tutti i numeri passati come argomenti.

```python
def print_kwargs(**kwargs):
    print(f"Keyword arguments: {kwargs}")
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_kwargs(name="Alice", age=30, city="Paris", email="alice@example.com")
```
