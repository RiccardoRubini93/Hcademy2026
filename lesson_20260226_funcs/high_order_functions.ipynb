funzioni che prendono in input altre funzioni

```python
#use case : voglio creare una funzione che applica una trasformazione arbitraria ad una lista

nums = [1,2,3,4,5,6]

def trasformation(data,my_func):
    return [my_func(n) for n in data]


func = lambda x:x + 5 
squared = trasformation(nums,func)
print(squared)
```
