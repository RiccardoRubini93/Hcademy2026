### map

applica una funzione a tutti gli elementi di una lista o di un iterable

```python
names = [" Andrea ","  Luca","Ric "]

list(map(lambda x:x.strip(),names))
```

### filter

```python
numbs = [1,2,3,4,5,6]

even_number = list(filter(lambda x:x%2==0,numbs))

even_number
```

### reduce

applica la lambda in modo iterativo alla lista 

```python
from functools import reduce

nums = [1,2,3,4,5]

sum = reduce(lambda x,y:x+y,numbs)

sum
```
