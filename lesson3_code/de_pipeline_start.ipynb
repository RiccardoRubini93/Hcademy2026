```python
# 1. DATA INPUT (A list of JSON-like dictionaries)
# Added duplicate records (ID 1 and ID 4) to demonstrate de-duplication
raw_orders = [
    {"order_id": 1, "customer": "Alice", "amount": 150.50, "status": "shipped"},
    {"order_id": 2, "customer": "Bob", "amount": None, "status": "shipped"},
    {"order_id": 1, "customer": "Alice", "amount": 150.50, "status": "shipped"}, 
    {"order_id": 3, "customer": "Charlie", "amount": -50.0, "status": "shipped"},
    {"order_id": 4, "customer": "David", "amount": "300.00", "status": "shipped"},
    {"order_id": 5, "customer": "Eve", "amount": 20.0, "status": "cancelled"},
    {"order_id": 4, "customer": "David", "amount": "300.00", "status": "shipped"},  
    {"order_id": 6, "customer": "Frank", "amount": 75.25, "status": "shipped"},
]

```

```python
# 2. INITIALIZATION -> Create empty lists to store valid orders, skipped orders, and duplicate orders
valid_amounts = []
skipped_orders = []
duplicate_orders = []
```

Apply the following rules:

1. DE-DUPLICATION CHECK ->  If the ID is already in our set, it's a duplicate
2. Business Rule: Only process 'shipped' orders
3. Data Cleansing -> amount > 0 e mount != None
4. KPI CALCULATION -> total_revenue, number_of_valid_orders
5. Create a report such as metrics {total_revenue,avg_value,unique_orders_success} audit {total_record_receive,dupicates_removed,skipped_invalid_orders,unique_id_seen}

```python
#rimozione dei duplicati
seen_order = set()  # To track seen order IDs for de-duplication

for order in raw_orders:  
    current_id = order["order_id"]

    if current_id in seen_order:
        print(f"Duplicate order found: {order}")
        duplicate_orders.append(order)
        continue  # Skip processing this duplicate order


    seen_order.add(current_id)  # Mark this order ID as seen
```

```python
#applicazione businesss rules-> considero solo ordini shipped  

good_orders = []

for order in raw_orders:

    current_id = order["order_id"]

    if order['status'] == 'shipped': 

        good_orders.append(current_id)
        print(f"selected order {current_id} due to status: {order['status']}")
        continue

    
```

```python
#implementazione step di data cleansing -> rimuover i None, rimuover i valori negativi e vogliamo assicuranrci che gli amount siano numerici. 

for order in raw_orders: 

    current_id = order["order_id"] 

    amount = order['amount']
    print(f"Processing order {current_id} with amount: {amount}")

    if amount is None: 
        skipped_orders.append(order)

    elif isinstance(amount, str):
        print(f"Amount is a string for order {current_id}, attempting to convert.")
        amount = float(amount)  # Convert string to float
        
        if amount > 0 : valid_amounts.append(amount)
        
        else: skipped_orders.append(order)

    elif amount > 0: valid_amounts.append(amount)

    else: 
        print(f"Negative amount found for order {current_id}, skipping.")
        skipped_orders.append(order)    
```

```python
total_revenue = sum(valid_amounts)
```

```python
count_valid = len(good_orders) 
```

```python

if count_valid > 0:
    avg_order_amt = total_revenue / count_valid
else:
    avg_order_amt = 0
```

```python
report = {
    "metrics" : {
        "total_revenue": total_revenue,
         "average_order_amount": avg_order_amt},

    "logs" : {
        "skipped_orders": len(skipped_orders),
        "good_orders":len(good_orders),
        "duplicate_orders": len(duplicate_orders)
    }  
}
```
