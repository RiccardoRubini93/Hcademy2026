# Task

you need build a function that compute the sum of the first N integers

you are also asked to optimise the function as much as possible 

```python
#use the following snippet of code to compute the performances of the code

import timeit

N = 10**6

# Timing the performance of both functions
time_v0 = timeit.timeit(lambda: sum_naive(N), number=5)
time_v1 = timeit.timeit(lambda: sum_pb(N), number=5)

print(f"Execution time (for loop version): {time_v0:.4f} seconds")
print(f"Execution time (set version): {time_v1:.4f} seconds")
print(f"Speedup: {time_v0 / time_v1:.2f}x")
```

```python
def sum_naive(n):
    sum = 0
    for i in range(1, n + 1):
        sum += i
    return sum

def sum_pb(n):
    return sum(range(1, n + 1))
```
