```python
def my_transfomrm(data,func):
    return [func(x) for x in data]
```

```python
func = lambda x:x**2
my_transfomrm([1,2,3],func)
```

```python
def greet():
    return print("Hello, World!")
```

```python
greet()
```

```python
def my_decorator(my_func):
    
    def wrap(): 
        
        print("Something is happening before the function is called.")
        my_func()
        print("Something is happening after the function is called.")
        
    return wrap 
```

```python
greet_decorated = my_decorator(greet) 
```

```python
greet_decorated()
```

```python
@my_decorator
def my_new_greet():
    return print("Hello, World!")
```

```python
greet()
```

```python
my_new_greet()
```

```python
def add(x,y):
    return x+y
```

```python
def my_log(func):

    def wrap(*args,**kwargs): 
        print(f"Executing {func.__name__} function")
        print(f"Function called with {args} and {kwargs} " )
        result = func(*args,**kwargs)
        return result
    return wrap 
```

```python
@my_log
def add_v2(x,y):
    return x+y
```

```python
add(1,2)
```

```python
add_v2(1,2)
```
