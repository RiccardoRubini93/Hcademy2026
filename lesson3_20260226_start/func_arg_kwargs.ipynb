```python
def create_connection_string(db_type, **params):
    """Genera una stringa di connessione dinamica."""
    if db_type == "postgres":
        return f"postgresql://{params.get('user')}:{params.get('password')}@{params.get('host')}:{params.get('port')}/{params.get('db')}"
    
    elif db_type == "snowflake":
        return f"snowflake://{params.get('account')}.snowflakecomputing.com/?warehouse={params.get('warehouse')}"
    
    elif db_type == "mysql":
        return f"mysql://{params.get('user')}:{params.get('password')}@{params.get('host')}:{params.get('port')}/{params.get('db')}"
    else:
        return "Tipo di DB non supportato"
```

```python
create_connection_string("postgres", password="admin",user='riccardo',host='local',port='1.2.3.4',db='mydb')
```

```python
create_connection_string("snowflake", account="myaccount", warehouse="mywarehouse")
```
